```python
import folium
from folium.plugins import MousePosition
from PyQt5 import QtCore, QtWidgets, QtWebEngineWidgets, QtWebChannel
import sys
import time
from qt_material import apply_stylesheet
import numpy as np
import clipboard
import import_ipynb
import ApoKomb1
import db1
import main_p
import GuidedLocalSearch
import GraphNeuralNetwork

import json
import io 


class Window(QtWidgets.QMainWindow):
    def __init__(self):
        super().__init__()
        self.result = None
        self.coords = []
        self.num_cities = 0
        self.markers = []

        self.initial_text_edit = QtWidgets.QTextEdit(self)
        self.initial_text_edit.setReadOnly(True)
        self.initial_text_edit.setFixedHeight(100)
        self.initial_text_edit.setFixedWidth(120)

        self.new_order_text_edit = QtWidgets.QTextEdit(self)
        self.new_order_text_edit.setReadOnly(True)
        self.new_order_text_edit.setFixedHeight(100)
        self.new_order_text_edit.setFixedWidth(120)

        self.time_text_edit = QtWidgets.QTextEdit(self)
        self.time_text_edit.setReadOnly(True)
        self.time_text_edit.setFixedHeight(100)
        self.time_text_edit.setFixedWidth(120)

        self.tsp_text_edit = QtWidgets.QTextEdit(self)
        self.tsp_text_edit.setReadOnly(True)
        self.tsp_text_edit.setFixedHeight(100)
        self.tsp_text_edit.setFixedWidth(180)
        
        self.layout_initialized = False

        self.initWindow()

    def initWindow(self):
        self.setWindowTitle("MAP OF GREECE")
        self.resize(1600, 1000)
        self.buttonUI()

    def buttonUI(self):
        points = [5, 10, 20, 50]
        self.view = QtWebEngineWidgets.QWebEngineView()
        self.view.setContentsMargins(50, 50, 50, 50)

        central_widget = QtWidgets.QWidget()
        self.setCentralWidget(central_widget)
        main_layout = QtWidgets.QHBoxLayout(central_widget)

        button_container = QtWidgets.QWidget()
        vlay = QtWidgets.QVBoxLayout(button_container)
        vlay.setSpacing(20)
        vlay.addStretch()
        
        self.add_button = QtWidgets.QPushButton("Add City")
        self.add_button.clicked.connect(self.add_closest_city)#
        self.add_button.setFixedSize(120, 30)
        vlay.addWidget(self.add_button)
        
        for num_points in points:
            button = QtWidgets.QPushButton(str(num_points))
            button.clicked.connect(lambda ch, num=num_points: self.handle_main_program(num))
            button.setFixedSize(120, 30)
            vlay.addWidget(button)

        self.distance_button = QtWidgets.QPushButton("Calc Dic")
        self.distance_button.clicked.connect(self.calculate_distance)
        self.distance_button.setFixedSize(120, 50)
        self.distance_button.setEnabled(False)
        vlay.addWidget(self.distance_button)
        
        self.clear_button = QtWidgets.QPushButton("Clear")
        self.clear_button.clicked.connect(self.clear_map)
        self.clear_button.setFixedSize(120, 50)
        vlay.addWidget(self.clear_button)
        
        vlay.addStretch()
        main_layout.addWidget(button_container, stretch=1)

        map_container = QtWidgets.QWidget()
        map_layout = QtWidgets.QVBoxLayout(map_container)
        map_layout.addWidget(self.view, stretch=100)

        text_area_container = QtWidgets.QWidget()
        text_area_layout = QtWidgets.QHBoxLayout(text_area_container)
        text_area_layout.addWidget(QtWidgets.QLabel("Initial Cities Order:"))
        text_area_layout.addWidget(self.initial_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("New Order of Cities:"))
        text_area_layout.addWidget(self.new_order_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("Time:"))
        text_area_layout.addWidget(self.time_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("GNN vs TSP_lib:"))
        text_area_layout.addWidget(self.tsp_text_edit)
        text_area_container.setLayout(text_area_layout)
        
        scroll_area = QtWidgets.QScrollArea()
        scroll_area.setWidgetResizable(True)
        scroll_area.setWidget(text_area_container)
        map_layout.addWidget(scroll_area)

        main_layout.addWidget(map_container, stretch=5)

        self.init_map()



    def init_map(self):
        self.m = folium.Map(location=[35, 25], zoom_start=5)
        self.m.add_child(folium.LatLngPopup())

        self.m.add_child(
            folium.ClickForLatLng(format_str='"[" + lat + "," + lng + "]"', alert=True)
        )
        
        
        
        
        #self.m.add_child(
            #folium.ClickForMarker("<b>Lat:</b> ${lat}<br /><b>Lon:</b> ${lng}"))
        
        data = io.BytesIO()
        self.m.save()

        with open('map.html', 'r') as f:
            html = f.read()

        
     
        
        self.view.setHtml(html)

 
    def handle_main_program(self, num_points):
        print(self.coords)
        self.result = main_p.basiko.main_program(self, num_points)
        self.node_citys, self.coords, self.edge_index, self.distances = main_p.basiko.city_names_and_coords(self.result)
        self.selected_cities = []  # Reset selected cities
        self.cities_coords = np.array(self.coords)
        self.num_cities = num_points
        initial_order_text = "\n".join(self.node_citys)
        self.display_initial_order(initial_order_text)
        self.distance_button.setEnabled(len(self.coords) >= 3)

    def calculate_distance(self):
        if self.result is not None:
            start_time = time.time()
            self.model_graph, self.edge_index_lg, self.node_features = main_p.graph_to_LineGraph.create_graph(self.edge_index, self.distances)

            dx = 1
            dh = 120
            output_dim = 1
            num_layers = 5

            self.distances_matrix = ApoKomb1.apostasis_test(self.node_citys, self.coords)
            self.model_gnn = GraphNeuralNetwork.GNNModel(dx, dh, output_dim, num_layers)

            self.regret_out = self.model_gnn(self.node_features, self.edge_index_lg)

            self.edge_index_n, self.regret_values = GuidedLocalSearch.Guided_local_Search.make_edge_index_and_regret(self.regret_out, self.edge_index)

            self.best_solution, self.best_dist = GuidedLocalSearch.Guided_local_Search.gls(self.distances_matrix, self.regret_values, self.edge_index_n, max_iterations=20)

            main_p.basiko.map_path(self)
            new_order = [self.node_citys[i] for i in self.best_solution]
            new_order_text = "\n".join(new_order)
            self.display_new_order(new_order_text)
            
            self.tsp_solver = GuidedLocalSearch.Guided_local_Search.tsp_solver(self.distances_matrix)
            
            tsp_result_text = f"Best Distance: {self.best_dist}\nTSP Solver: {self.tsp_solver}"
            self.tsp_new_order(tsp_result_text)
            
            end_time = time.time()
            time_taken = end_time - start_time
            self.display_time_taken(f"{time_taken:.2f} seconds")
            
        else:
            print("No result to calculate distance from.")
            
    
    def add_closest_city(self):
        
        self.coords= clipboard.paste()
        print(self.coords) 
        
        
    def clear_map(self):
        self.coords = []
        self.m = folium.Map(location=[35, 25], zoom_start=5)
        self.m.add_child(folium.LatLngPopup())
        
        self.m.add_child(
            folium.ClickForLatLng(format_str='"[" + lat + "," + lng + "]"', alert=False)
        )
        
        
        self.coords=clipboard.paste()
        
        
        #self.m.add_child(
            #folium.ClickForMarker("<b>Lat:</b> ${lat}<br /><b>Lon:</b> ${lng}"))
        self.m.save('map.html')

        with open('map.html', 'r') as f:
            html = f.read()

        self.view.setHtml(html)
        
        self.distance_button.setEnabled(False)

    def OnomataPolewn(self):
        if not self.layout_initialized:
            self.layout_initialized = True

    def display_initial_order(self, text):
        self.initial_text_edit.setPlainText(text)

    def display_new_order(self, text):
        self.new_order_text_edit.setPlainText(text)
        
    def display_time_taken(self, text):
        self.time_text_edit.setPlainText(text)
        
    def tsp_new_order(self, text):
        self.tsp_text_edit.setPlainText(text)
        
        
    

if __name__ == "__main__":
    try:
        App = QtWidgets.QApplication(sys.argv)
        apply_stylesheet(App, theme='dark_teal.xml')
        window = Window()
        window.show()
        sys.exit(App.exec())
    except Exception as e:
        print("Exception:", e)
        sys.exit(1)
```

```python
import sys
import time
import numpy as np
import folium
from folium.plugins import MousePosition
from PyQt5 import QtCore, QtWidgets, QtWebEngineWidgets, QtWebChannel
from qt_material import apply_stylesheet
import clipboard
import import_ipynb
import ApoKomb1
import db1
import main_p
import GuidedLocalSearch
import GraphNeuralNetwork
#import map

class Window(QtWidgets.QMainWindow):
    def __init__(self):
        super().__init__()
        self.result = None
        self.coords = []
        self.num_cities = 0
        self.markers = []

        self.initial_text_edit = QtWidgets.QTextEdit(self)
        self.initial_text_edit.setReadOnly(True)
        self.initial_text_edit.setFixedHeight(100)
        self.initial_text_edit.setFixedWidth(120)

        self.new_order_text_edit = QtWidgets.QTextEdit(self)
        self.new_order_text_edit.setReadOnly(True)
        self.new_order_text_edit.setFixedHeight(100)
        self.new_order_text_edit.setFixedWidth(120)

        self.time_text_edit = QtWidgets.QTextEdit(self)
        self.time_text_edit.setReadOnly(True)
        self.time_text_edit.setFixedHeight(100)
        self.time_text_edit.setFixedWidth(120)

        self.tsp_text_edit = QtWidgets.QTextEdit(self)
        self.tsp_text_edit.setReadOnly(True)
        self.tsp_text_edit.setFixedHeight(100)
        self.tsp_text_edit.setFixedWidth(180)

        self.layout_initialized = False
        self.selected_lat_lng = None

        self.initWindow()

    def initWindow(self):
        self.setWindowTitle("MAP OF GREECE")
        self.resize(1900, 1000)
        #self.showMaximized()
        self.buttonUI()

    def buttonUI(self):
        points = [5, 10, 20, 50]
        self.view = QtWebEngineWidgets.QWebEngineView()
        self.view.setContentsMargins(50, 50, 50, 50)

        central_widget = QtWidgets.QWidget()
        self.setCentralWidget(central_widget)
        main_layout = QtWidgets.QHBoxLayout(central_widget)

        button_container = QtWidgets.QWidget()
        vlay = QtWidgets.QVBoxLayout(button_container)
        vlay.setSpacing(20)
        vlay.addStretch()
        
        #self.button_test = QtWidgets.QPushButton("Sotiris")
        #self.button_test.clicked.connect({getLatLng})
        #self.button_test.setFixedSize(120, 30)
        
        for num_points in points:
            button = QtWidgets.QPushButton(str(num_points))
            button.clicked.connect(lambda ch, num=num_points: self.handle_main_program(num))
            button.setFixedSize(120, 30)
            vlay.addWidget(button)

        self.add_button = QtWidgets.QPushButton("Add")
        self.add_button.clicked.connect(self.add_coordinates)
        self.add_button.setFixedSize(120, 50)
        vlay.addWidget(self.add_button)

        self.distance_button = QtWidgets.QPushButton("Calc Dic")
        self.distance_button.clicked.connect(self.calculate_distance)
        self.distance_button.setFixedSize(120, 50)
        self.distance_button.setEnabled(False)
        vlay.addWidget(self.distance_button)

        self.clear_button = QtWidgets.QPushButton("Clear")
        self.clear_button.clicked.connect(self.clear_map)
        self.clear_button.setFixedSize(120, 50)
        vlay.addWidget(self.clear_button)

        vlay.addStretch()
        main_layout.addWidget(button_container, stretch=1)

        map_container = QtWidgets.QWidget()
        map_layout = QtWidgets.QVBoxLayout(map_container)
        map_layout.addWidget(self.view, stretch=100)

        text_area_container = QtWidgets.QWidget()
        text_area_layout = QtWidgets.QHBoxLayout(text_area_container)
        text_area_layout.addWidget(QtWidgets.QLabel("Initial Cities Order:"))
        text_area_layout.addWidget(self.initial_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("New Order of Cities:"))
        text_area_layout.addWidget(self.new_order_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("Time:"))
        text_area_layout.addWidget(self.time_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("GNN vs TSP_lib:"))
        text_area_layout.addWidget(self.tsp_text_edit)
        text_area_container.setLayout(text_area_layout)

        scroll_area = QtWidgets.QScrollArea()
        scroll_area.setWidgetResizable(True)
        scroll_area.setWidget(text_area_container)
        map_layout.addWidget(scroll_area)

        main_layout.addWidget(map_container, stretch=5)

        self.init_map()

    def init_map(self):
        self.m = folium.Map(location=[35, 25], zoom_start=5)
        self.m.add_child(folium.LatLngPopup())

       
        

        #self.m.save('map.html')

        with open('map.html', 'r') as f:
            html = f.read()

        self.view.setHtml(html)
        self.view.page().loadFinished.connect(self.add_js_interaction)

    def add_js_interaction(self):
        js_script = """
        document.addEventListener('DOMContentLoaded', function() {
            var map = document.getElementsByClassName('folium-map')[0];
            if (map) {
                map._leaflet_map.on('click', function(e) {
                    var lat = e.latlng.lat;
                    var lng = e.latlng.lng;
                    new QWebChannel(qt.webChannelTransport, function(channel) {
                        var pyObj = channel.objects.pyObj;
                        pyObj.handle_click(lat.toString() + ',' + lng.toString());
                    });
                    console.log('Coordinates sent to PyQt:', lat, lng);
                });
            }
        });
        """

        channel = QtWebChannel.QWebChannel(self.view.page())
        self.view.page().setWebChannel(channel)
        channel.registerObject("pyObj", self)
        self.view.page().runJavaScript(js_script)

    @QtCore.pyqtSlot(str)
    def handle_click(self, latlng):
        lat, lng = map(float, latlng.split(','))
        self.selected_lat_lng = (lat, lng)
        print(f"Coordinates selected: {lat}, {lng}")

    def add_coordinates(self):
        if self.selected_lat_lng:
            lat, lng = self.selected_lat_lng
            self.coords.append((lat, lng))
            print(f"Coordinates added: {lat}, {lng}")
            if len(self.coords) >= 3:
                self.distance_button.setEnabled(True)
            print(f"Total coordinates: {len(self.coords)}")
        else:
            print("No coordinates selected.")

    def handle_main_program(self, num_points):
        print(self.coords)
        self.result = main_p.basiko.main_program(self, num_points)
        self.node_citys, self.coords, self.edge_index, self.distances = main_p.basiko.city_names_and_coords(self.result)
        self.selected_cities = []  # Reset selected cities
        self.cities_coords = np.array(self.coords)
        self.num_cities = num_points
        initial_order_text = "\n".join(self.node_citys)
        self.display_initial_order(initial_order_text)
        self.distance_button.setEnabled(len(self.coords) >= 3)

    def calculate_distance(self):
        if self.result is not None:
            start_time = time.time()
            self.model_graph, self.edge_index_lg, self.node_features = main_p.graph_to_LineGraph.create_graph(self.edge_index, self.distances)

            dx = 1
            dh = 120
            output_dim = 1
            num_layers = 5

            self.distances_matrix = ApoKomb1.apostasis_test(self.node_citys, self.coords)
            self.model_gnn = GraphNeuralNetwork.GNNModel(dx, dh, output_dim, num_layers)

            self.regret_out = self.model_gnn(self.node_features, self.edge_index_lg)

            self.edge_index_n, self.regret_values = GuidedLocalSearch.Guided_local_Search.make_edge_index_and_regret(self.regret_out, self.edge_index)

            self.best_solution, self.best_dist = GuidedLocalSearch.Guided_local_Search.gls(self.distances_matrix, self.regret_values, self.edge_index_n, max_iterations=20)

            main_p.basiko.map_path(self)
            new_order = [self.node_citys[i] for i in self.best_solution]
            new_order_text = "\n".join(new_order)
            self.display_new_order(new_order_text)

            self.tsp_solver = GuidedLocalSearch.Guided_local_Search.tsp_solver(self.distances_matrix)

            tsp_result_text = f"Best Distance: {self.best_dist}\nTSP Solver: {self.tsp_solver}"
            self.tsp_new_order(tsp_result_text)

            end_time = time.time()
            time_taken = end_time - start_time
            self.display_time_taken(f"{time_taken:.2f} seconds")

        else:
            print("No result to calculate distance from.")

    def clear_map(self):
        self.coords = []
        self.init_map()
        self.distance_button.setEnabled(False)

    def OnomataPolewn(self):
        if not self.layout_initialized:
            self.layout_initialized = True

    def display_initial_order(self, text):
        self.initial_text_edit.setPlainText(text)

    def display_new_order(self, text):
        self.new_order_text_edit.setPlainText(text)

    def display_time_taken(self, text):
        self.time_text_edit.setPlainText(text)

    def tsp_new_order(self, text):
        self.tsp_text_edit.setPlainText(text)

if __name__ == "__main__":
    try:
        App = QtWidgets.QApplication(sys.argv)
        apply_stylesheet(App, theme='dark_teal.xml')
        window = Window()
        window.show()
        sys.exit(App.exec())
    except Exception as e:
        print("Exception:", e)
        sys.exit(1)
```

```python
import folium
from folium.plugins import MousePosition
from PyQt5 import QtCore, QtWidgets, QtWebEngineWidgets, QtWebChannel
import sys
import time
from qt_material import apply_stylesheet
import numpy as np

import import_ipynb
import ApoKomb1
import db1
import main_p
import GuidedLocalSearch
import GraphNeuralNetwork


class Window(QtWidgets.QMainWindow):
    def __init__(self):
        super().__init__()
        self.result = None
        self.coords = []
        self.num_cities = 0

        self.initial_text_edit = QtWidgets.QTextEdit(self)
        self.initial_text_edit.setReadOnly(True)
        self.initial_text_edit.setFixedHeight(100)
        self.initial_text_edit.setFixedWidth(120)

        self.new_order_text_edit = QtWidgets.QTextEdit(self)
        self.new_order_text_edit.setReadOnly(True)
        self.new_order_text_edit.setFixedHeight(100)
        self.new_order_text_edit.setFixedWidth(120)

        self.time_text_edit = QtWidgets.QTextEdit(self)
        self.time_text_edit.setReadOnly(True)
        self.time_text_edit.setFixedHeight(100)
        self.time_text_edit.setFixedWidth(120)

        self.tsp_text_edit = QtWidgets.QTextEdit(self)
        self.tsp_text_edit.setReadOnly(True)
        self.tsp_text_edit.setFixedHeight(100)
        self.tsp_text_edit.setFixedWidth(180)

        self.initWindow()

    def initWindow(self):
        self.setWindowTitle("MAP OF GREECE")
        self.resize(1600, 1000)
        self.buttonUI()

    def buttonUI(self):
        points = [5, 10, 20, 50]
        self.view = QtWebEngineWidgets.QWebEngineView()
        self.view.setContentsMargins(50, 50, 50, 50)

        central_widget = QtWidgets.QWidget()
        self.setCentralWidget(central_widget)
        main_layout = QtWidgets.QHBoxLayout(central_widget)

        button_container = QtWidgets.QWidget()
        vlay = QtWidgets.QVBoxLayout(button_container)
        vlay.setSpacing(20)
        vlay.addStretch()
        
        #self.Add_button = QtWidgets.QPushButton("Add")
        #self.Add_button.clicked.connect(self.handle_click([35,20]))
        #self.Add_button.setFixedSize(120, 50)
        #vlay.addWidget(self.Add_button)

        for num_points in points:
            button = QtWidgets.QPushButton(str(num_points))
            button.clicked.connect(lambda ch, num=num_points: self.handle_main_program(num))
            button.setFixedSize(120, 30)
            vlay.addWidget(button)

        self.distance_button = QtWidgets.QPushButton("Calc Dic")
        self.distance_button.clicked.connect(self.calculate_distance)
        self.distance_button.setFixedSize(120, 50)
        self.distance_button.setEnabled(False)
        vlay.addWidget(self.distance_button)

        vlay.addStretch()
        main_layout.addWidget(button_container, stretch=1)

        map_container = QtWidgets.QWidget()
        map_layout = QtWidgets.QVBoxLayout(map_container)
        map_layout.addWidget(self.view, stretch=100)

        text_area_container = QtWidgets.QWidget()
        text_area_layout = QtWidgets.QHBoxLayout(text_area_container)
        text_area_layout.addWidget(QtWidgets.QLabel("Initial Cities Order:"))
        text_area_layout.addWidget(self.initial_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("New Order of Cities:"))
        text_area_layout.addWidget(self.new_order_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("Time:"))
        text_area_layout.addWidget(self.time_text_edit)
        text_area_layout.addWidget(QtWidgets.QLabel("GNN vs TSP_lib:"))
        text_area_layout.addWidget(self.tsp_text_edit)
        text_area_container.setLayout(text_area_layout)
        
        scroll_area = QtWidgets.QScrollArea()
        scroll_area.setWidgetResizable(True)
        scroll_area.setWidget(text_area_container)
        map_layout.addWidget(scroll_area)

        main_layout.addWidget(map_container, stretch=5)

        self.init_map()

    def init_map(self):
        m = folium.Map(location=[35, 25], zoom_start=5)
        m.add_child(folium.LatLngPopup())

        formatter = "function(num) {return L.Util.formatNum(num, 5);};"
        mouse_position = MousePosition(
            position='topright',
            separator=' Long: ',
            empty_string='Unavailable',
            lng_first=True,
            num_digits=5,
            prefix='Lat:',
            lat_formatter=formatter,
            lng_formatter=formatter,
        )
        m.add_child(mouse_position)
        m.add_child(folium.ClickForMarker(popup="Marker"))
        m.save('map.html')

        with open('map.html', 'r') as f:
            html = f.read()
        self.view.setHtml(html)

        # Set up the web channel for communication
        channel = QtWebChannel.QWebChannel()
        self.view.page().setWebChannel(channel)
        channel.registerObject("pyObj", self)
        self.view.page().runJavaScript("""
            function handleMapClick(event) {
                var latlng = event.latlng;
                var lat = latlng.lat;
                var lng = latlng.lng;
                pyObj.handle_click(lat + ',' + lng);
                console.log('Coordinates sent to PyQt:', lat, lng);
            }

            new QWebChannel(qt.webChannelTransport, function(channel) {
                var pyObj = channel.objects.pyObj;
                var map = document.getElementsByClassName('leaflet-container')[0]._leaflet_map;
                map.on('click', handleMapClick);
            });
        """)

    def on_click_map(self):
        m = folium.Map(location=[35, 25], zoom_start=5)
        m.add_child(folium.LatLngPopup())

        formatter = "function(num) {return L.Util.formatNum(num, 5);};"
        mouse_position = MousePosition(
            position='topright',
            separator=' Long: ',
            empty_string='Unavailable',
            lng_first=True,
            num_digits=5,
            prefix='Lat:',
            lat_formatter=formatter,
            lng_formatter=formatter,
        )
        
        m.add_child(mouse_position)
        m.add_child(folium.ClickForMarker(popup="Marker"))
        m.save('map.html')

        with open('map.html', 'r') as f:
            html = f.read()
        self.view.setHtml(html)

        # Set up the web channel for communication
        channel = QtWebChannel.QWebChannel()
        self.view.page().setWebChannel(channel)
        channel.registerObject("pyObj", self)
        self.view.page().runJavaScript("""
            new QWebChannel(qt.webChannelTransport, function(channel) {
                var pyObj = channel.objects.pyObj;
                var map = document.getElementsByClassName('leaflet-container')[0]._leaflet_map;
                map.on('click', function(event) {
                    var latlng = event.latlng;
                    var lat = latlng.lat;
                    var lng = latlng.lng;
                    pyObj.handle_click(lat + ',' + lng);
                    console.log('Coordinates sent to PyQt:', lat, lng);
                });
            });
        """)    
    

    @QtCore.pyqtSlot(str)
    
    def handle_click(self, latlng):
        #lat, lng = map(float, latlng.split(','))
        #self.coords.append((lat, lng))
        self.coords.append((latlng[0],latlng[1]))
        #print(f"Coordinates added: {lat}, {lng}")  # Debug print to verify the coordinates are added
        if len(self.coords) >= 3:
            self.distance_button.setEnabled(True)
        print(f"Total coordinates: {len(self.coords)}")  # Additional debug print
        #on_click_map()

    def handle_main_program(self, num_points):
        self.result = main_p.basiko.main_program(self, num_points)
        self.node_citys, self.coords, self.edge_index, self.distances = main_p.basiko.city_names_and_coords(self.result)
        self.selected_cities = []  # Reset selected cities
        self.cities_coords = np.array(self.coords)
        self.num_cities = num_points
        initial_order_text = "\n".join(self.node_citys)
        self.display_initial_order(initial_order_text)
        self.distance_button.setEnabled(len(self.coords) >= 3)

    def calculate_distance(self):
        if self.result is not None:
            start_time = time.time()
            self.model_graph, self.edge_index_lg, self.node_features = main_p.graph_to_LineGraph.create_graph(self.edge_index, self.distances)

            dx = 1
            dh = 120
            output_dim = 1
            num_layers = 5

            self.distances_matrix = ApoKomb1.apostasis_test(self.node_citys, self.coords)
            self.model_gnn = GraphNeuralNetwork.GNNModel(dx, dh, output_dim, num_layers)

            self.regret_out = self.model_gnn(self.node_features, self.edge_index_lg)

            self.edge_index_n, self.regret_values = GuidedLocalSearch.Guided_local_Search.make_edge_index_and_regret(self.regret_out, self.edge_index)

            self.best_solution, self.best_dist = GuidedLocalSearch.Guided_local_Search.gls(self.distances_matrix, self.regret_values, self.edge_index_n, max_iterations=20)

            main_p.basiko.map_path(self)
            new_order = [self.node_citys[i] for i in self.best_solution]
            new_order_text = "\n".join(new_order)
            self.display_new_order(new_order_text)

            self.tsp_solver = GuidedLocalSearch.Guided_local_Search.tsp_solver(self.distances_matrix)

            tsp_result_text = f"Best Distance: {self.best_dist}\nTSP Solver: {self.tsp_solver}"
            self.tsp_new_order(tsp_result_text)

            end_time = time.time()
            time_taken = end_time - start_time
            self.display_time_taken(f"{time_taken:.2f} seconds")
        else:
            print("No result to calculate distance from.")

    def display_initial_order(self, text):
        self.initial_text_edit.setPlainText(text)

    def display_new_order(self, text):
        self.new_order_text_edit.setPlainText(text)

    def display_time_taken(self, text):
        self.time_text_edit.setPlainText(text)

    def tsp_new_order(self, text):
        self.tsp_text_edit.setPlainText(text)

if __name__ == "__main__":
    try:
        App = QtWidgets.QApplication(sys.argv)
        apply_stylesheet(App, theme='dark_teal.xml')
        window = Window()
        window.show()
        sys.exit(App.exec())
    except Exception as e:
        print("Exception:", e)
        sys.exit(1)
```

```python
import import_ipynb
from PyQt5 import QtCore, QtGui, QtWidgets, QtWebEngineWidgets
import folium
import ApoKomb1
import db1
import warnings
import main_p
import GuidedLocalSearch
import GraphNeuralNetwork
import time

import io
import sys
from qt_material import apply_stylesheet

warnings.simplefilter(action='ignore', category=FutureWarning)

class Window(QtWidgets.QMainWindow):

    def __init__(self):
        super().__init__()
        self.result = None  # Instance variable to hold the result from main_program

        # Initialize text edits here
        self.initial_text_edit = QtWidgets.QTextEdit(self)
        self.initial_text_edit.setReadOnly(True)
        self.initial_text_edit.setFixedHeight(80)
        
        self.new_order_text_edit = QtWidgets.QTextEdit(self)
        self.new_order_text_edit.setReadOnly(True)
        self.new_order_text_edit.setFixedHeight(80)
        
        self.time_text_edit = QtWidgets.QTextEdit(self)
        self.time_text_edit.setReadOnly(True)
        self.time_text_edit.setFixedHeight(50)
        
        self.tsp_text_edit = QtWidgets.QTextEdit(self)
        self.tsp_text_edit.setReadOnly(True)
        self.tsp_text_edit.setFixedHeight(50)
        
        self.initWindow()
        self.layout_initialized = False
        
    def initWindow(self):
        self.setWindowTitle(self.tr("MAP OF GREECE"))
        self.resize(1600, 1000)
        self.buttonUI()

    def buttonUI(self):
        points = [5, 10, 20, 50]
        self.view = QtWebEngineWidgets.QWebEngineView()
        self.view.setContentsMargins(50, 50, 50, 50)

        central_widget = QtWidgets.QWidget()
        self.setCentralWidget(central_widget)
        main_layout = QtWidgets.QHBoxLayout(central_widget)  # Changed to QHBoxLayout

        button_container = QtWidgets.QWidget()
        vlay = QtWidgets.QVBoxLayout(button_container)
        vlay.setSpacing(20)
        vlay.addStretch()

        for num_points in points:
            button = QtWidgets.QPushButton(self.tr(str(num_points)))
            button.clicked.connect(lambda ch, num=num_points: self.handle_main_program(num))
            button.setFixedSize(120, 30)
            vlay.addWidget(button)
            
        distance_button = QtWidgets.QPushButton(self.tr("Calc Dic"))
        distance_button.clicked.connect(self.calculate_distance)
        distance_button.setFixedSize(120, 50)
        vlay.addWidget(distance_button)
        
        vlay.addStretch()
        main_layout.addWidget(button_container)
        
        map_layout = QtWidgets.QVBoxLayout()
        map_layout.addWidget(self.view, stretch=1)

        m = folium.Map(location=[35, 25], zoom_start=5)
        data = io.BytesIO()
        m.save(data, close_file=False)
        self.view.setHtml(data.getvalue().decode(), QtCore.QUrl('file://'))

        # Add the text areas below the map
        map_layout.addWidget(QtWidgets.QLabel("Initial Cities Order:"))
        map_layout.addWidget(self.initial_text_edit)
        map_layout.addWidget(QtWidgets.QLabel("New Order of Cities:"))
        map_layout.addWidget(self.new_order_text_edit)
        map_layout.addWidget(QtWidgets.QLabel("Time Taken for Calculations:"))
        map_layout.addWidget(self.time_text_edit)
        map_layout.addWidget(QtWidgets.QLabel("Oi apostash apo to modelo vs TSP library solver:"))
        map_layout.addWidget(self.tsp_text_edit)
        
        main_layout.addLayout(map_layout)

    def handle_main_program(self, num_points):
        self.result = main_p.basiko.main_program(self, num_points)
        self.node_citys, self.coords, self.edge_index, self.distances = main_p.basiko.city_names_and_coords(self.result)
        initial_order_text = "\n".join(self.node_citys)
        self.display_initial_order(initial_order_text)
        self.OnomataPolewn()

    def calculate_distance(self):
        if self.result is not None:
            start_time = time.time()
            self.model_graph, self.edge_index_lg, self.node_features = main_p.graph_to_LineGraph.create_graph(self.edge_index, self.distances)

            dx = 1
            dh = 120
            output_dim = 1
            num_layers = 5

            self.distances_matrix = ApoKomb1.apostasis_test(self.node_citys, self.coords)
            self.model_gnn = GraphNeuralNetwork.GNNModel(dx, dh, output_dim, num_layers)

            self.regret_out = self.model_gnn(self.node_features, self.edge_index_lg)

            self.edge_index_n, self.regret_values = GuidedLocalSearch.Guided_local_Search.make_edge_index_and_regret(self.regret_out, self.edge_index)

            self.best_solution, self.best_dist = GuidedLocalSearch.Guided_local_Search.gls(self.distances_matrix, self.regret_values, self.edge_index_n, max_iterations=20)

            main_p.basiko.map_path(self)
            new_order = [self.node_citys[i] for i in self.best_solution]
            new_order_text = "\n".join(new_order)
            self.display_new_order(new_order_text)
            
            self.tsp_solver = GuidedLocalSearch.Guided_local_Search.tsp_solver(self.distances_matrix)
            
            tsp_result_text = f"Best Distance: {self.best_dist}\nTSP Solver: {self.tsp_solver}"
            self.tsp_new_order(tsp_result_text)
            
            end_time = time.time()
            time_taken = end_time - start_time
            self.display_time_taken(f"{time_taken:.2f} seconds")
            
        else:
            print("No result to calculate distance from.")
            
    def OnomataPolewn(self):
        if not self.layout_initialized:
            self.layout_initialized = True

    def display_initial_order(self, text):
        self.initial_text_edit.setPlainText(text)

    def display_new_order(self, text):
        self.new_order_text_edit.setPlainText(text)
        
    def display_time_taken(self, text):
        self.time_text_edit.setPlainText(text)
        
    def tsp_new_order(self, text):
        self.tsp_text_edit.setPlainText(text)

if __name__ == "__main__":
    try:
        App = QtWidgets.QApplication(sys.argv)
        apply_stylesheet(App, theme='dark_teal.xml')  # Apply Qt Material theme here
        window = Window()
        window.show()
        sys.exit(App.exec())
    except Exception as e:
        print("Exception:", e)
        sys.exit(1)
```
